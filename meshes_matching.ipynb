# Matching meshes with $ MGW_2 $

This notebook shows how to use $ MGW_2 $ to asses a distance between two meshes

```python
import numpy as np
import matplotlib.pyplot as plt
from MGW.mgw2 import MGW2
```

First let load the data.

```python
meshes = np.load('data/horses_meshes.npy')

X = meshes[0]
Y = meshes[9]
```

```python
plt.scatter(X[:,2],X[:,1],cmap='Blues',c=X[:,0],s=0.3)
plt.show()
```

```python
plt.scatter(Y[:,2],Y[:,1],cmap='Blues',c=Y[:,0],s=0.3)
plt.show()
```

Now compute $ MGW_2 $ between the two meshes.

```python
n_components = 20


dist = MGW2(X,Y,n_components=n_components)
```
